```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MyLR:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):
        X = X_train.values.flatten()   # convert to 1D array
        y = y_train.values

        x_mean = np.mean(X)
        y_mean = np.mean(y)

        num = 0
        den = 0

        for i in range(len(X)):
            num += (X[i] - x_mean) * (y[i] - y_mean)
            den += (X[i] - x_mean) ** 2

        self.m = num / den
        self.b = y_mean - (self.m * x_mean)

        print("Slope (m):", self.m)
        print("Intercept (b):", self.b)
        print(x_mean)
        print(y_mean)

    def predict(self, X_test):
        X = X_test.values.flatten()
        return self.m * X_test + self.b


# Load dataset
df = pd.read_csv("cgpa_data.csv")

X = df[["CGPA"]]
Y = df["Package_LPA"]

# Split FIRST
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# Create object
lr = MyLR()

# Train
lr.fit(X_train, Y_train)

# Predict
predictions = lr.predict(X_test)

print(predictions[:5])
```
